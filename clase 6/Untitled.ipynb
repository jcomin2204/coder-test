```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo
datos = {'País': ['Rusia', 'Canadá', 'Estados Unidos', 'China', 'Brasil', 'Australia', 'India', 'Argentina'],
         'Superficie (km²)': [17098242, 9984670, 9833517, 9596961, 8515767, 7692024, 3287263, 2780400]}
df = pd.DataFrame(datos)

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))  # Establecer el tamaño del gráfico
plt.bar(df['País'], df['Superficie (km²)'], color='skyblue')

# Añadir etiquetas y título
plt.xlabel('País')
plt.ylabel('Superficie (km²)')
plt.title('Superficie de los 8 países más grandes del mundo')

# Rotar las etiquetas del eje x para mejor visualización
plt.xticks(rotation=45, ha='right')

# Mostrar el gráfico
plt.tight_layout()  # Ajustar diseño
plt.show()
```
